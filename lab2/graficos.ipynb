```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
robo = pd.read_csv('arqSaida.csv')
```

```python
robo.columns
```

### Gráfico u(t)

```python

plt.figure(figsize=(10, 6))
plt.plot(robo['t'], robo['v'], label='Velocidade Linear', color='b')


plt.title('Parte v(t) - velocidade linear')
plt.xlabel('Tempo (t)')
plt.ylabel('Valores de v')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(robo['t'], robo['w'], label='Velocidade Angular', color='r')


plt.title('Parte w(t) - velocidade Angular')
plt.xlabel('Tempo (t)')
plt.ylabel('Valores de w')
plt.legend()
plt.show()
```

### Grafico y(t)

```python
plt.figure(figsize=(10, 6))
plt.plot(robo['t'], robo['y1'], label='y1 - Xc', color='g')


plt.title('Parte y1(t) - Xc')
plt.xlabel('Tempo (t)')
plt.ylabel('Valores de y1')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(robo['t'], robo['y2'], label='y1 - Yc', color='g')


plt.title('Parte y2(t) - Yc')
plt.xlabel('Tempo (t)')
plt.ylabel('Valores de y2')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(robo['t'], robo['y3'], label='y3 - Teta', color='g')


plt.title('Parte y3(t) - Teta')
plt.xlabel('Tempo (t)')
plt.ylabel('Valores de y3')
plt.legend()
plt.show()
```
