```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
robo = pd.read_csv('arqSaida.csv')
```

```python
robo.columns
```

### Gráfico y(k)

```python

plt.figure(figsize=(10, 6))
plt.plot(robo['xref'], robo['yref'], label='Referencia', color='b')


plt.title('Referencia')
plt.xlabel('Xref')
plt.ylabel('Yref')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(robo['x1'], robo['x2'], label='Trajetoria', color='r')


plt.title('Trajetoria')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

 - Período

```python
T = []
for i in range(1, len(robo['t'])):
    T.append((robo['t'][i] - robo['t'][i-1])*1000) # em ms

len(T)
```

```python
T = pd.DataFrame(T)
```

```python
# media 
Tmedia = np.mean(T)
Tmedia
```

```python
# variancia
np.var(T)
```

```python
# desvio padrão
np.std(T)
```

```python
# maximo
np.max(T)
```

```python
# minimo
np.min(T)
```

- Jitter

```python
Delta_T = T - Tmedia
Delta_T
```

```python
Jitter = np.mean(np.abs(Delta_T))
Jitter
```
